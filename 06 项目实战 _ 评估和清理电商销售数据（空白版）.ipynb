# 项目：评估和清理英国电商公司销售数据

## 分析目标

此数据分析的目的是，根据市场销售数据，挖掘畅销产品，以便制定更有效的市场策略来提升营收。

本实战项目的目的在于练习评估数据干净和整洁度，并且基于评估结果，对数据进行清洗，从而得到可供下一步分析的数据。

## 简介

原始数据集记录了一家英国在线零售公司在2010年12月1日至2011年12月9日期间的所有交易情况，涵盖了该公司在全球不同国家和地区的业务数据。该公司主要销售覆盖各个场景的礼品，包括但不限于生日礼品、结婚纪念品、圣诞礼品等等。该公司的客户群体主要包括批发商和个人消费者，其中批发商占据了相当大的比例。

数据每列的含义如下：
- `InvoiceNo`: 发票号码。6位数，作为交易的唯一标识符。如果这个代码以字母“c”开头，表示这笔交易被取消。
- `StockCode`: 产品代码。5位数，作为产品的唯一标识符。
- `Description`: 产品名称。
- `Quantity`: 产品在交易中的数量。
- `InvoiceDate`: 发票日期和时间。交易发生的日期和时间。
- `UnitPrice`: 单价。价格单位为英镑（£）。
- `CustomerID`: 客户编号。5位数，作为客户的唯一标识符。
- `Country`: 国家名称。客户所居住的国家的名称。

## 读取数据

```python
import pandas as pd

data = pd.read_csv('e_commerce.csv')
data.sample(10)
```

## 评估数据

```python
data.info()
# Description  CustomerID   包含缺失值
# InvoiceDate  CustomerID   类型错误
```

```python
# Description包含缺失值 且 对应的UnitPrice都为0 可以删除
# CustomerID为用户ID不影响分析 可以保留   
data[data['Description'].isnull() & data['UnitPrice'] != 0]
```

```python
# 不一致数据
data['Country'].value_counts()
# 修改多个名称代表同一国家的情况
```

## 清理数据

```python
# 获得原始数据备份
mark_data = data
```

```python
# CustomerID类型为float 需要转为字符串
mark_data['CustomerID'] = mark_data['CustomerID'].astype('str')
# InvoiceDate 类型为object 需要转换为日期
mark_data['InvoiceDate'] = pd.to_datetime(data['InvoiceDate'])
mark_data.info()
```

```python
mark_data.sample(10)
# 这里在CustomerID变为字符串后结尾有.0需要删除
# data['CustomerID'].str.slice(0,-2) 
mark_data['CustomerID'] = mark_data['CustomerID'].str.replace('.0', '')
mark_data
```

## 【自己想的回来解决】 试着加入确实Description 对应的唯一 StockCode存在于数据中 能否填充

```python
mark_data[mark_data['CustomerID'].isnull()]
# 这里出现意外 由于前边将该列转为字符串格式 缺失值消失了 变为了字符串'nan'
mark_data[mark_data['CustomerID']=='nan']
```

```python
# 删除Description缺失值
mark_data.dropna(subset='Description', inplace=True)
# 检查
mark_data['Description'].isnull().sum()
# 检查2
mark_data.info()
```

```python
# USA --> United States UK --> United Kingdom U.K. --> United Kingdom
mark_data['Country'] = mark_data['Country'].replace({'USA': 'United States', 'UK': 'United KingdomKingdom','U.K.': 'United Kingdom',})
mark_data['Country'].value_counts()
```

```python
# 删除Quantity < 0, UnitPrice < 0
mark_data = mark_data[mark_data['Quantity'] >= 0]
mark_data = mark_data[mark_data['UnitPrice'] >= 0]
```
